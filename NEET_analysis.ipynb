#### 2. Ingest the data file and format into a Pandas DataFrame

```python
df = pd.DataFrame(columns=['rank', 'quota', 'institute', 'city', 'state', 'course', 'category', 'candidate', 'remarks'])

with open("round_1.txt", "r") as file:
    i = 1
    for line in file:
#         if i == 200:
#             break
        if line.replace("\n", "").isdecimal():
            rank = file.readline().replace("\n", "")
            quota = file.readline().replace("\n", "")
            institute_state = file.readline().replace("\n", "").split(',')
            institute = institute_state[0]
            city = "" if len(institute_state) == 1 else institute_state[1]
            course = file.readline().replace("\n", "")
            category = file.readline().replace("\n", "")
            candidate = file.readline().replace("\n", "")
            remarks = file.readline().replace("\n", "")
            df.loc[i] = [rank,quota,institute,city,"",course,category,candidate,remarks]
            i += 1
```
